```python
# Cargamos paqueterías
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python

# Catgamos datos de Default (se pueden conseguir en paquete ILSR de R)
url = "https://raw.githubusercontent.com/jasonm/islr-exercises/refs/heads/master/Data/Default.csv"
df = pd.read_csv(url)
# Convertimos las binarias a numéricas
df["default"] = df["default"].map({"No": 0, "Yes": 1})
df["student"] = df["student"].map({"No": 0, "Yes": 1})
X = df[["student", "balance", "income"]]
y = df["default"]
```

```python
# Separamos datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Preprocesamos los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Modelo de regresión logística con penalización de red elástica
logreg = LogisticRegression(
    penalty="elasticnet",
    solver="saga",
    max_iter=5000
)
# Malla de parámetros para validación cruzada
param_grid = {
    "C": [0.01, 0.1, 1, 10, 20, 50, 100],
    "l1_ratio": [0.0, 0.25, 0.5, 0.75, 1.0]  # 0=L2, 1=L1
}
# Validación cruzada
logregCV = GridSearchCV(
    logreg,
    param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)
```
